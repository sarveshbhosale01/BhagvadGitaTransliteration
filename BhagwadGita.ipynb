Data Preprocessing

```python
data.drop(columns={'Verse'},inplace=True) #Dropping the column
```

```python
import re
def clean_id(text):
  return re.sub(r'[a-zA-Z]','',text)
clean_id('BG1.1')
```

```python
data['ID'] = data['ID'].apply(clean_id) #Cleaning operation
```

Shloka Cleaning function

```python
import re

def clean_Shloka(text):
    # Remove pipes and colons
    text = re.sub(r'[|:]', '', text)

    # Remove visarga (ः) if needed
    text = re.sub(r'ः', '', text)

    # Remove numerals at end like ||१-१||
    text = re.sub(r'\|*\d+-\d+\|*', '', text)
    text = re.sub(r'\|*[\d१२३४५६७८९०]+-[\d१२३४५६७८९०]+\|*', '', text)  # for Devanagari numerals

    # Remove double danda || if left
    text = re.sub(r'\|\|', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
data['Shloka'] = data['Shloka'].apply(clean_Shloka)
```

Transliteration Cleaning function

```python
def clean_Transliteration(text):
   text = re.sub(r'\|+\d+-\d+\|+',' ',text)
   text = re.sub(r'\.\s','',text)
   text = re.sub(r'\.','',text)
   return text
```

```python
data['Transliteration'] = data['Transliteration'].apply(clean_Transliteration)
```

HindiMeaning Column Cleaning

```python
def clean_HinMeaning(text):
   text = re.sub(r'।।\d+\.\d+।।', '', text)
   text = re.sub(r'-','',text)
   text = re.sub(r'\|','',text)
   text = re.sub(r'\(.*?\)','',text)
   text = re.sub(r'\!','',text)

   return text
```

```python
data['HinMeaning'] = data['HinMeaning'].apply(clean_HinMeaning)
```

EnglishMeaning Column Cleaning

```python
def clean_EngMeaning(text):
   text = re.sub(r'\d+\.\d','', text)
   text = re.sub(r'[\n,.""?-]','',text)
   text = re.sub(r'\d+',' ',text)
   text = re.sub(r'(\a-zA-Z)','',text)
   text = re.sub(r'\([^)]*\)', '', text)
   return text
```

```python
data['EngMeaning'] = data['EngMeaning'].apply(clean_EngMeaning)
```

WordMeaning Cleaning function

```python
def clean_WordMeaning(text):
   text = re.sub(r'\d+\.+\d','',text)
   text = re.sub(r'\d','',text)
   text = re.sub(r'[.,?*--:]','',text)
   text = re.sub(r'\([^)]*\)','',text)
   return text
```

```python
data['WordMeaning'] = data['WordMeaning'].apply(clean_WordMeaning)
```

Main System

```python
def find_shloka_meaning(data: pd.DataFrame, shloka: str) -> dict or None:
    try:
        row = data[data["Shloka"] == shloka].iloc[0]
        result = {
            "Transliteration": row["Transliteration"],
            "HinMeaning": row["HinMeaning"],
            "EngMeaning": row["EngMeaning"],
            "WordMeaning": row["WordMeaning"],
        }
        return result
    except IndexError:
        return None
input_shloka = "यत्र योगेश्वर कृष्णो यत्र पार्थो धनुर्धर तत्र श्रीर्विजयो भूतिर्ध्रुवा नीतिर्मतिर्मम"
result = find_shloka_meaning(data, input_shloka)

if result:
    print(f"For the shloka: '{input_shloka}'")
    print(f"  Transliteration: {result['Transliteration']} \n")
    print(f"  Hindi Meaning: {result['HinMeaning']} \n")
    print(f"  English Meaning: {result['EngMeaning']} \n")
    print(f"  Word Meaning: {result['WordMeaning']} \n")
else:
    print(f"Shloka '{input_shloka}' not found in the data.")

    input_shloka_not_found = "This shloka is not present."
    result_not_found = find_shloka_meaning(data, input_shloka_not_found)
    if result_not_found is None:
        print(f"\nShloka '{input_shloka_not_found}' not found in the data (as expected).")
```
